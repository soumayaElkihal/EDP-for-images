```python
from tkinter import *
import tkinter.messagebox
import tkinter.filedialog
from tkinter import simpledialog


import PIL
from PIL import Image,ImageTk
import numpy as np
import math
from scipy import misc
import cv2
import matplotlib.pyplot as plt 
from tkinter.filedialog import *
import tkinter.messagebox
import tkinter.filedialog
def test():
    global filename
imag_result='None'






############################################ l'interface grafique ###############################
####################################### ### ##############################################
Mafenetre = Tk()
Mafenetre.title("Traitement d'images sous python") 
#Mafenetre.config( bg='grey')
Mafenetre.config( background='#F1F9C7')
#variable
Mafenetre.iconbitmap(r"C:\Users\21266\downloads\sosss.ico")





# Création d'un widget Menu
menubar = Menu(Mafenetre)

menufichier = Menu(menubar,tearoff=0)
menufichier.add_command(label="Ouvrir une image",command=lambda:Ouvrir()) 
menufichier.add_command(label="Enregistrer l'image",command=lambda:save_image() )


menubar.add_cascade(label="image", menu=menufichier)

menuP = Menu(menubar,tearoff=0)
menuP.add_command(label="A propos",command= lambda:Apropos())
menubar.add_cascade(label="A propos", menu=menuP)




# Affichage du menu
Mafenetre.config(menu=menubar)
      


#################### creation des frames ###########################################################
titre = Frame(Mafenetre,width=1600,height=50 ,bg='#F1F9C7', relief=GROOVE)
titre.pack(side=TOP)
f0 = Frame(Mafenetre,width=1600,height=50 ,bg='#F1F9C7', relief=GROOVE)
f0.pack(side=TOP)

f1 = Frame(Mafenetre,width=250,height=600 ,bg='#E7F0B7', relief=GROOVE)
f1.pack(side=LEFT)

f2 = Frame(Mafenetre,width=800,height=600 ,bg='#E7F0B7', relief=GROOVE)
f2.pack(side=LEFT)

f3 = Frame(Mafenetre,width=250,height=600 ,bg='#E7F0B7', relief=GROOVE)
f3.pack(side=LEFT)

#icon=Canvas(titre)
#icon.pack(side=RIGHT )


########################################### remplir les frames ###################################
labtit = Label(titre,font=('arial',25,'bold'), text="Traitement d'images sous python",bg='#F1F9C7' ,fg='black', bd=10, anchor='w')
labtit.pack()
#PP=PIL.Image.open('Icon.png')
#photo = ImageTk.PhotoImage(image=PP)
#icon.create_image(0,0,anchor=NW,image=photo)### afficher logo
#icon.config(height=photo.height(),width=photo.width())

###################################### les filtres ########################
l1= Label(f0,font=('arial',20,'bold'), text="Filtres :",bg='#F1F9C7', fg='black', bd=10, anchor='center')
l1.grid(row=0,column=2)
b1= Button(f0,font=('arial',10,'bold '),bg='lavender', text ="Moyenne",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :Moyenne() ).grid(row=1,column=0)
b2= Button(f0,font=('arial',10,'bold '),bg='lavender', text ="Binomial",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :Binomial()).grid(row=1,column=1)
b3= Button(f0,font=('arial',10,'bold '),bg='lavender', text ="Gauss",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :Gauss()).grid(row=1,column=2)
b4= Button(f0, font=('arial',10,'bold '),bg='lavender',text ="Log",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :Log()).grid(row=1,column=3)
b5= Button(f0,font=('arial',10,'bold '), bg='lavender',text ="Dog",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :Dog()).grid(row=1,column=4)


############################ detection des coutours ####################################
l2= Label(f1,font=('arial',20,'bold'), text="Detection  des contours :",bg='#F1F9C7', fg='black', bd=10, anchor='center')
l2.grid(row=0,columnspan=2 )

b1= Button(f1,font=('arial',10,'bold '), bg='lavender',text ="directionnel",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :directionnel()).grid(row=1)
b2= Button(f1,font=('arial',10,'bold '), bg='lavender',text ="Robert",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:Robert()).grid(row=2)
b3= Button(f1, font=('arial',10,'bold '),bg='lavender',text ="Sobel",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:Sobel()).grid(row=3)
b4= Button(f1,font=('arial',10,'bold '), bg='lavender',text ="Prewitt",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:Prewit()).grid(row=4)
b5= Button(f1, text ="Kirsh",width=15,font=('arial',10,'bold '),bg='lavender', height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:Krish()).grid(row=1,column=1)
b6= Button(f1,font=('arial',10,'bold '),bg='lavender', text ="Laplacien",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:Laplatien()).grid(row=4,column=1)
b7= Button(f1,font=('arial',15,'bold ') ,bg='lavender',text ="Tracer contours",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:win_trace()).grid(row=9,columnspan=2)



######################################### les EDP #############################################
l3= Label(f3,font=('Times',20,'bold'), text="Réstauration par EDP :",bg='#F1F9C7', fg='black', bd=10, anchor='center')
l3.grid(row=0,column=1,columnspan=2)
b1= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="Chaleur",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda :chaleur() ).grid(row=1,column=1)
b2= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="Chaleur réaction 1",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:reaction() ).grid(row=2,column=1)
b3= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="Chaleur réaction 2",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:reaction_ameliore()).grid(row=3,column=1)
b4= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="Perona-Malik 1",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:malik()).grid(row=4,column=1)
b5= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="Perona-Malik 2",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:malik2()).grid(row=5,column=1)
b6= Button(f3,font=('arial',10,'bold '), bg='lavender',text ="PM Convolution 1",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:malik_conv()).grid(row=3,column=2)
b9= Button(f3, font=('arial',10,'bold '),bg='lavender',text ="PM Convolution 2",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:malik_conv1()).grid(row=4,column=2)




   





# Création d'un widget Canvas pour afficher l'images 

w=Canvas(f2, bg='beige')
w.pack(padx=5,pady=5)
b1= Button(f2, font=('arial',10,'bold '),bg='lavender',text ="Image initial",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE, command=lambda:initial() ).pack(side=RIGHT)
b2= Button(f2, font=('arial',10,'bold '),bg='lavender',text ="Image résultat",width=15, height=1,borderwidth=5,fg='black', relief=GROOVE,command=lambda:result() ).pack(side=LEFT)
label = Label(f2,font=('arial',10,'bold '), text='Acune opération ',fg='red',bg='lavender',width=20, height=1,borderwidth=5,relief=FLAT)
label.pack()




################################ enregistrer l'image resultat #########################
def save_image():
    global label
    global imag_result
    if imag_result == 'None':
        tkinter.messagebox.showwarning('Attention!!!', "Pas d'image pour l'instant")
    else:   
        file_name=asksaveasfilename(title=" enregistrer l'image",defaultextension='.png ',filetypes=[('Images','.png '),('all files','.*')])#enregistrer l'image
        cv2.imwrite(file_name,miroir_extract(imag_result))
    label.config(text='Image enregistrer')
######################### redimentioner image pour l'affichage #####################
def redimention(imag):
    height,width=imag.size
    if (height, width) > (600,450):
        imag=imag.resize((600,450))
    return imag    
#################### ouvrire image ###########################
def Ouvrir():
    # on efface la zone graphique
    global filename
    global imag_initial 
    global imag_initial1 
    global imag_result
    global label
    filename = tkinter.filedialog.askopenfilename(title="Ouvrir une image",filetypes=[('Images','.png .jpg'),('all files','.*')])
    #print(filename)
    imag_initial =miroir_expand(cv2.imread(filename,0))
    imag_initial1 =imag_initial/255
    
    imag_result=imag_initial
    photo = tab_photo(imag_initial)
    gifdict[filename] = photo  # référence
    #print(photo)
    
    Affiche(photo)
    #imag_initial=cv2.imread(filename,0)
    label.config(text='Image initial')

############################### afficher image dans la place réserver ####################
    

def Affiche(photo):
    w.delete(ALL) 
    gifdict[filename] = photo
    w.create_image(0,0,anchor=NW,image=photo)
    w.config(height=photo.height(),width=photo.width())
    Mafenetre.title("Image "+str(photo.width())+" x "+str(photo.height()))
    
##################### transformer tableau au image ###############
def tab_photo(tab):
    
    
    P=PIL.Image.fromarray(miroir_extract(tab))
    P=redimention(P)
    photo = ImageTk.PhotoImage(image=P)
    return photo
    #P.show()

####################### les deux button afficher image initial et resultat##################
def initial():
    global label
    photo=tab_photo(imag_initial)
    Affiche(photo)
    label.config(text='Image initial')
def result():
    photo=tab_photo(imag_result)
    Affiche(photo)
    label.config(text='Image résultat')

######################################################################    
def Apropos():
    tkinter.messagebox.showinfo("SOUMIA'S APPLICATION","Cette application est réalisé par  SOUMAYA EL KIHAL  \n Etudiante en  licence  (MIASI). \n FSTG Marrakech .  ")

######################################## les miroirs #################################""
def   miroir_extract(B):
    m,n = np.shape(B) 
    A = np.zeros((m-2,n-2))
    for x in range(0,m-2):
        for y in range(0,n-2):
   
           
            A[x,y] = B[x+1,y+1]
    return(A)



def  miroir_expand(A):
    
    m,n = np.shape(A) 
    B = np.zeros((m+2,n+2))
    for x in range(1,m+1):
        for y in range(1,n+1):
   
           
            B[x,y] = A[x-1,y-1]
           
            B[0 ,y] = B[2 ,y]
            B[m+1,y] = B[m-1,y]
            B[x,0] = B[x,2]
            B[x,n+1] = B[x,n-1] 
            B[0,0] = B[2,2]
            B[0,n+1] = B[2,n-1]
            B[m+1,0] = B[m-1,2]
            B[m+1,n+1] = B[m-1,n-1]
    return (B)

def Convolution2D(Imag,Filtre):
    n,m=np.shape(Imag)
    p0 = np.zeros((n,m))
    for x in range(1,n-1):
    
        for y in range(1,m-1):
            

                     V=Imag[x-1:x+2,y-1:y+2] 
                     p0[x,y] = (np.sum(V*Filtre))
        
    return (p0)

def laplacien(B):
    n,m=np.shape(B)
    p=np.zeros((n,m))
    D=np.array(([1, 1, 1],[1 ,-8 ,1],[1, 1, 1]))
    for x in range(1,n-1):
    
        for y in range(1,m-1):
            V=B[x-1:x+2,y-1:y+2]
            aide = V*D
            p[x,y] = (np.sum(aide))

    return(p)

#filtre


############################### fenitre dialogue pour les filtres  ##############################
def new_win():
    global nomb_fois
    global sigma
    def recupere():
        global nomb_fois
        global sigma
        
        nomb_fois=int(e1.get())  
        sigma=float(e2.get() )
        win1.destroy() 
    global b1
    global value
    
    win1 = Toplevel(Mafenetre)
    value = StringVar()
    value1 = StringVar()
    Label(win1,text=" Nombre de fois à appliquer ce filtre :").grid(row=0) 
    Label(win1,text=" Sigma :").grid(row=1) 
    
    
    e1 = Entry(win1, textvariable=value , width=30)
    e1.grid(row=0,column=1) 
    
    e2 = Entry(win1, textvariable=value1 , width=30)
    e2.grid(row=1,column=1) 
    
    
    b1 = Button(win1,text='validé', command=lambda:recupere())
    b1.grid(row=3,column=2) 
    
    
    Mafenetre.wait_window(win1)

def new_win1():
    global nomb_fois
    global sigma
    global sigma1
    def recupere():
        global nomb_fois
        global sigma
        global sigma1
        
        nomb_fois=int(e1.get() )  
        sigma=float(e2.get())
        sigma1=float(e3.get())
        win2.destroy() 
    
    
    win2 = Toplevel(Mafenetre)
    value = StringVar()
    value1 = StringVar()
    value2 = StringVar()
    Label(win2,text=" Nombre de fois à appliquer ce filtre :").grid(row=0) 
    Label(win2,text=" Sigma1 :").grid(row=1) 
    Label(win2,text=" Sigma2 :").grid(row=2) 
    
    e1 = Entry(win2, textvariable=value , width=30)
    e1.grid(row=0,column=1) 
    
    
    e2 = Entry(win2, textvariable=value1 , width=30)
    e2.grid(row=1,column=1) 
    
    e3 = Entry(win2, textvariable=value2 , width=30)
    e3.grid(row=2,column=1) 
    
    b1 = Button(win2,text='validé', command=lambda:recupere())
    b1.grid(row=4,column=2) 
    
    
    Mafenetre.wait_window(win2)    
    
    

    
############################  Filtres  ###############################    
    
def Moyen(Imag):
    P=Convolution2D(Imag,np.array(([1,1,1],[1,1,1],[1,1,1]))/9)
    
    return P
def Binom(Imag):
    P=Convolution2D(Imag,np.array(([1,2,1],[2,4,2],[1,2,1]))/16)
    
    return P



def gauss(Imag,sg):
    ssg=sg**2
    b=np.exp(-1/ssg)
    bb=np.sqrt(b)
    d=2*math.pi*ssg
    G=np.array(([b, bb ,b],[bb ,1 ,bb],[b, bb, b]))
    G=G/d
    G=G/np.sum(G)
    P=Convolution2D(Imag,G)
    
    return P
def log(Imag,sg):
    ssg=sg**2
    b=np.exp(-1/ssg)
    bb=np.sqrt(b)
    d=2*math.pi*ssg
    G0=np.array(([b, bb ,b],[bb ,1 ,bb],[b, bb, b]))
    G1=np.array(([2*b, bb ,2*b],[bb ,0 ,bb],[2*b, bb, 2*b]))
    G = 2*G1- 2*ssg*G0
    G=G/d
    G=G/np.sum(G)
    P=Convolution2D(Imag,G)
    
    return P
def dog(Imag,sg,sg1):
    ssg=sg**2
    b=np.exp(-1/ssg)
    bb=np.sqrt(b)
    G0=np.array(([b, bb ,b],[bb ,1 ,bb],[b, bb, b]))/2*math.pi*ssg
    
    ssg1=sg1**2
    b1=np.exp(-1/ssg1)
    bb1=np.sqrt(b1)
    G1=np.array(([b1, bb1 ,b1],[bb1 ,1 ,bb1],[b1, bb1, b1]))/2*math.pi*ssg1
    G=G0-G1
    G=G/np.sum(G)
    P=Convolution2D(Imag,G)
    
    return P
    
    
   
    



def Moyenne():
    
    
    global label
    nombre_de_fois=tkinter.simpledialog.askinteger('Entré nombre de fois pour appliquer ce filtre','N',parent=Mafenetre)
    
   
    global imag_result
    global imag_initial
    tab=imag_initial
    for j in range(nombre_de_fois):
        tab1=Moyen(tab)
        tab=tab1
    
    imag_result=tab 
    photo=tab_photo(tab)
    Affiche(photo)
    label.config(text='Filtre moyenne')
    
    
    
    
    
def Binomial():
    nombre_de_fois=tkinter.simpledialog.askinteger('Entré nombre de fois pour appliquer ce filtre','N',parent=Mafenetre)
    global label
    global imag_result
    global imag_initial
    tab=imag_initial
    for j in range(nombre_de_fois):
        tab1=Binom(tab)
        tab=tab1
        
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Filtre binomial')
def Gauss():
    global nomb_fois
    global sigma
    global label

    new_win()
    
    global imag_result
    global imag_initial
    tab=imag_initial
    for j in range(nomb_fois):
        tab1=gauss(tab,sigma)
        tab=tab1
        
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Filtre gaussien')

    
def Dog():
    global nomb_fois
    global sigma
    global sigma1
    global label
    new_win1()
    global imag_result
    global imag_initial
    tab=imag_initial
    for j in range(nomb_fois):
        tab1=dog(tab,sigma,sigma1)
        tab=tab1
        
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Filtre Dog')
    
def Log():
    global nomb_fois
    global sigma
    global label

    new_win()
    
    global imag_result
    global imag_initial
    tab=imag_initial
    for j in range(nomb_fois):
        tab1=log(tab,sigma)
        tab=tab1
        
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Filtre Log')
#############################################les contours#########################################


def Robert():
    global label
    global image_cont
    #derive directionnel  
    global imag_result
    global imag_initial
    X=np.array(([0, 0, 0],[0, 0, 1],[0, -1, 0]))
    Y=np.array(([0, 0 ,0],[0 ,1 ,0],[0, 0, -1]))
    dX=Convolution2D(imag_initial,X)
    dY=Convolution2D(imag_initial,Y)
    tab=np.sqrt(dX**2+dY**2)
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)    
    label.config(text='Mask de Robert')
def Sobel():
    global label
    global image_cont
    global imag_result
    global imag_initial
    X=np.array(([-1,-2 , -1],[0, 0, 0],[1, 2, 1]))/4
    Y=np.array(([-1, 0 ,1],[-2 ,0 ,2],[-1, 0, 1]))/4
    dX=Convolution2D(imag_initial,X)
    dY=Convolution2D(imag_initial,Y)
    tab=np.sqrt(dX**2+dY**2)
    imag_result=tab
    image_cont=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)  
    label.config(text='Mask de sobel')
def Prewit():
    global label
    global image_cont
    global imag_result
    global imag_initial
    X=np.array(([-1, -1, -1],[0, 0, 0],[1, 1, 1]))/3
    Y=np.array(([-1, 0 ,1],[-1 , 0,1],[-1, 0, 1]))/3
    dX=Convolution2D(imag_initial,X)
    dY=Convolution2D(imag_initial,Y)
    tab=np.sqrt(dX**2+dY**2)
    imag_result=tab
    image_cont=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo) 
    label.config(text='Mask de prewitt')
def Krish():
    global label
    global image_cont
    global imag_result
    global imag_initial
    X=np.array(([-3, -3, 5],[-3, 0, 5],[-3, -3, 5]))/15
    Y=np.array(([5 ,-3, -3],[5, 0 ,-3],[5, -3, -3]))/15
    dX=Convolution2D(imag_initial,X)
    dY=Convolution2D(imag_initial,Y)
    tab=np.sqrt(dX**2+dY**2)
    imag_result=tab
    image_cont=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo) 
    label.config(text='mask de kirsh')
    
 
def Laplatien():
    global label
    global image_cont
    global imag_result
    global imag_initial
    tab=laplacien(imag_initial)
    image_cont=tab
    imag_result=tab
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo) 
    label.config(text='Laplacien ')

    
def win_trace():
    global label
    global image_cont 
    global imag_result
    global imag_initial
    
    def Trace(B,degre):
        global imag_result
        n,m=np.shape(B)
        tab=np.zeros((n,m))
        for x in range(n):
            for y in range(m):
                if B[x,y]>degre:
                    tab[x,y]=0
                else:
                    tab[x,y]=255
        photo=tab_photo(tab)
        imag_result=tab
         
        gifdict[filename] = photo
        Affiche(photo) 
    def pres(x):
        return((10-x)*4+6)

    
    win3 = Toplevel(Mafenetre)
    
    Mafenetre.title(" Traces les contours en choisissans la precésion")
   
    s=Scale(win3, orient='horizontal', from_=0, to=100,
      resolution=10, tickinterval=10, length=350,
      label='Prcésion')
    s.pack()
    
    b1 = Button(win3,text='Validé', command=lambda:Trace(image_cont,pres(s.get() /10)))
    b1.pack()
    label.config(text='Tracer les contours')
    
    
    Mafenetre.wait_window(win3)     

############################################### EDP ####################################
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg #la trace des courbe psnr
from matplotlib.figure import Figure


def tracer(psnr):
    
        

    window=Toplevel(Mafenetre)


    fig = Figure(figsize=(6,6))
    a = fig.add_subplot(111)

    a.plot(psnr,color='blue')
#a.invert_yaxis()

    a.set_title ("PSNR", fontsize=16)
    a.set_ylabel("psnr", fontsize=14)
    a.set_xlabel("itération", fontsize=14)

    canvas = FigureCanvasTkAgg(fig, master=window)
    canvas.get_tk_widget().pack()
    canvas.draw()
    window.mainloop()

def PSNR(A,B):
    mse=np.mean((A-B)**2)
    if mse == 0:
        return 100
    else:
        return 10*math.log10(255**2/mse)
    
def grad(B):   
    (n,m)=np.shape(B)
    gradBx=np.zeros((n,m))
    gradBy=np.zeros((n,m))
    for i in range(1,n-1): 
        for j in range(1,m-1):
            gradBx[i,j] =B[i+1,j]-B[i,j]
            gradBy[i,j]=B[i,j+1]-B[i,j]
    return[gradBx,gradBy]
def f(u,b,a):
    # cette fonction pour la restauration par la chaleur
    #on définit les parramettres de la fonction f 
  
    return (b*u*(1-u)*(u-a))




def llambda(B):
    #la fonction lambda 
    [P1,P2]=grad(B)
    P=np.sqrt(P1**2+P2**2)
    M1=np.median(P)
    Q1=np.sqrt((P1-M1)**2+(P2-M1)**2)
    M2=np.median(Q1)
    s=1.4826*M2/np.sqrt(2)
    return s


def cngrad(B,Lambda):
    (Em,En)=np.shape(B)
    P=np.zeros((Em,En))
    [P1,P2]=grad(B)
    for i in range(1,Em-1):
        for j in range(1,En-1):
            q11=c(np.sqrt(P1[i,j]**2+P2[i,j]**2),Lambda)*P1[i,j]
            q22=c(np.sqrt(P1[i,j]**2+P2[i,j]**2),Lambda)*P2[i,j]
            q12=c(np.sqrt(P1[i-1,j]**2+P2[i-1,j]**2),Lambda)*P1[i-1,j] 
            q21=c(np.sqrt(P1[i,j-1]**2+P2[i,j-1]**2),Lambda)*P2[i,j-1] 
            P[i,j]=q11+q22-q12-q21
    return P
def c(r,Lambda):
    s=1/(1+(r/Lambda)**2)
    return s
def conv_gauss(B,sg):

    ssg=sg**2
    d=2*math.pi*ssg
    b=np.exp(-1/ssg)
    bb=np.exp(-1/(2*ssg)) 
    G=np.array(([b, bb ,b],[bb, 1, bb],[b, bb, b]))

    G=G/d 
    G=G/np.sum(G)
    p0=Convolution2D(B,G)

           
        
    return p0

def cngradm(B,sigma,Lambda):
    
    (Em,En)=np.shape(B)
    P=np.zeros((Em,En))
#calcul de P=div(c(norm(grad(G_sigma*u),Lambda).grad(u))
    BB=conv_gauss(B,sigma)
    [Q1,Q2]=grad(BB)
    [P1,P2]=grad(B) 
    for i in range(1,Em-1):
        for j in range(1,En-1):
            q11=c(np.sqrt(Q1[i,j]**2+Q2[i,j]**2),Lambda)*P1[i,j]
            q22=c(np.sqrt(Q1[i,j]**2+Q2[i,j]**2),Lambda)*P2[i,j]
            q12=c(np.sqrt(Q1[i-1,j]**2+Q2[i-1,j]**2),Lambda)*P1[i-1,j] 
            q21=c(np.sqrt(Q1[i,j-1]**2+Q2[i,j-1]**2),Lambda)*P2[i,j-1] 
            P[i,j]=q11+q22-q12-q21
    return P

################################ les fenitre dialoge pour EDP ########################
def chaleur_win():
    global T
    global dt
    
    def recupere():
        global T
        global dt
        
        
        T=float(e1.get() )  
        dt=float(e2.get())
        
        win2.destroy() 
    
    
    win2 = Toplevel(Mafenetre)
    value = StringVar()
    value1 = StringVar()
    
    Label(win2,text=" T =").grid(row=0) 
    Label(win2,text=" dt :").grid(row=1) 
     
    
    e1 = Entry(win2, textvariable=value , width=30)
    e1.grid(row=0,column=1) 
    
    
    e2 = Entry(win2, textvariable=value1 , width=30)
    e2.grid(row=1,column=1) 
    
    
    
    b1 = Button(win2,text='validé', command=lambda:recupere())
    b1.grid(row=4,column=2) 
    
    
    Mafenetre.wait_window(win2) 
    
def chaleur_win1():
    global T
    global dt
    global sigma
    def recupere():
        global T
        global dt
        global sigma
        
        
        T=float(e1.get() )  
        dt=float(e2.get())
        sigma=float(e3.get())
        
        win2.destroy() 
    
    
    win2 = Toplevel(Mafenetre)
    value = StringVar()
    value1 = StringVar()
    value2 = StringVar()
    Label(win2,text=" T =").grid(row=0) 
    Label(win2,text=" dt :").grid(row=1) 
    Label(win2,text=" sigma :").grid(row=2)  
     
    e1 = Entry(win2, textvariable=value , width=30)
    e1.grid(row=0,column=1) 
    
    
    e2 = Entry(win2, textvariable=value1 , width=30)
    e2.grid(row=1,column=1) 
    e3 = Entry(win2, textvariable=value2 , width=30)
    e3.grid(row=2,column=1) 
    
    
    
    b1 = Button(win2,text='validé', command=lambda:recupere())
    b1.grid(row=4,column=2) 
    
    
    Mafenetre.wait_window(win2)  

############################################# methodes EDP #############################
def chaleur():
    global label
    global imag_result
    global imag_initial1
    global dt
    global T
    chaleur_win()
    
    K=int(T/dt)
    psnr=np.zeros(K)
    B=imag_initial1
    for k in range(K):
        B=B+dt*laplacien(B)
        psnr[k] =PSNR(imag_initial1,B)
    
   
    tab=np.array(B*255,dtype=np.uint8)  
    
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo) 
    label.config(text='Chaleur')
    tracer(psnr)
    
    
    
def reaction():
    global label
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    K=int(T/dt)
    psnr=np.zeros(K)
    B=imag_initial1
    for k in range(K):
        B=B+dt*laplacien(B)+dt*f(B,2,0.25)
        psnr[k] =PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo) 
    label.config(text='Chaleur réaction 1')
    tracer(psnr)
def reaction_ameliore(): 
    global label
    tkinter.messagebox.showinfo("information", ' réaction en changeant a en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    K=int(T/dt)
    psnr=np.zeros(K)
    B=imag_initial1
    a=0.25*B
    for k in range(K):
        B=B+dt*laplacien(B)+dt*f(B,2,a)
        a=a+dt*laplacien(a)
        psnr[k] =PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8)  
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Chaleur réaction 2')
    tracer(psnr)
def malik():
    global label

    global imag_result
    global imag_initial1
    global dt
    global T
    chaleur_win()
    K=int(T/dt)
    psnr=np.zeros(K)
    Lambda = llambda(miroir_extract(imag_initial1))
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora malik 1')
    tracer(psnr)
def malik2():
    global label
    tkinter.messagebox.showinfo("information", 'malik en changeant lambda en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    chaleur_win()
    K=int(T/dt)

    B=imag_initial1
    Lambda=llambda(B)
    psnr=np.zeros(K)
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)
        psnr[j] =PSNR(imag_initial1,B)
        Lambda=llambda(B)
        

    tab=np.array(B*255,dtype=np.uint8) 
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora malik 2')
    tracer(psnr)
def malik_reaction():
    global label
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    K=int(T/dt)
    B=imag_initial1
    Lambda=llambda(miroir_extract(imag_initial1))
    
    psnr=np.zeros(K)
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)+dt*f(B,beta,0.25)
        psnr[j] =PSNR(imag_initial1,B)

    tab=np.array(B*255,dtype=np.uint8)  
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M réaction 1')
    tracer(psnr)
def malik_reaction1(): 
    global label
    tkinter.messagebox.showinfo("information", 'malik réaction en changeant a en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    a=0.25*imag_initial1
    
    K=int(T/dt)

    B=Imag
    Lambda=llambda(miroir_extract(B))
    L=np.zeros(K)
    psnr=np.zeros(K)
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)+dt*f(B,2,a)
        a=a+dt*laplacien(a)
        psnr[j] =PSNR(imag_initial1,B)


    tab=np.array(B*255,dtype=np.uint8)  
    
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M réaction 2')
    tracer(psnr)
    
def malik_reaction2():  
    global label
    tkinter.messagebox.showinfo("information", 'malik réaction en changeant lambda en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    
    K=int(T/dt)

    B=imag_initial1
    Lambda=llambda(B)
    psnr=np.zeros(K)
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)+dt*f(B,2,0.5)
        psnr[j] =PSNR(imag_initial1,B)
        Lambda=llambda(B)
    

    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M réaction 3')
    tracer(psnr)
def malik_reaction3():   
    global label
    tkinter.messagebox.showinfo("information", 'malik réaction en changeant lambda et a en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    
    chaleur_win()
    
    K=int(T/dt)

    B=imag_initial1
    a=0.25*B
    Lambda=llambda(B)
    psnr=np.zeros(K)
    for j in range(K):
        B=B+dt*cngrad(B,Lambda)+dt*f(B,2,a)
        a=a+dt*laplacien(a)
        psnr[j] =PSNR(imag_initial1,B)
        Lambda=llambda(B)
    

    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M réaction 4')
    tracer(psnr)        
def malik_conv(): 
    global label
    tkinter.messagebox.showinfo("information", 'Amélioration de Perona-Malik: Convolution du gradient avec une gaussienne')
    global imag_result
    global imag_initial1
    global dt
    global T
    global sigma
    
    chaleur_win1()
    K=int(T/dt)
    psnr=np.zeros(K)
    
    Lambda = llambda(miroir_extract(imag_initial1))
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngradm(B,sigma,Lambda)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M convolution 1')
    tracer(psnr)
def malik_conv1():  
    global label
    tkinter.messagebox.showinfo("information", 'Malik Convolué en changeant lambda et a en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    global sigma
    
    chaleur_win1()
    
    K=int(T/dt)
    psnr=np.zeros(K)
    
    Lambda = llambda(imag_initial1)
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngradm(B,sigma,Lambda)
        Lambda=llambda(B)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M convolution 2')
    tracer(psnr)
def malik_conv_reaction(): 
    global label
    tkinter.messagebox.showinfo("information", 'Malik Convolué: ajout d’un terme de réaction diffusion ')
    global imag_result
    global imag_initial1
    global dt
    global T
    global sigma
    chaleur_win1()
    
    K=int(T/dt)
    psnr=np.zeros(K)
    Lambda = llambda(miroir_extract(imag_initial1))
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngradm(B,sigma,Lambda)+dt*f(B,2,0.5)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M_C réaction 1')
    tracer(psnr)

def malik_conv_reaction1():
    global label
    tkinter.messagebox.showinfo("information", 'Malik Convolué: ajout d’un terme de réaction diffusion en changeant a en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    global sigma
    chaleur_win1()
    a=0.5*imag_initial1
    K=int(T/dt)
    psnr=np.zeros(K)
    Lambda = llambda(miroir_extract(imag_initial1))
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngradm(B,sigma,Lambda)+dt*f(B,2,a)
        a=a+dt*cngrad(a,Lambda)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M_C réaction 2')
    tracer(psnr)

def malik_conv_reaction2():
    global label 
    tkinter.messagebox.showinfo("information", 'Malik Convolué: ajout d’un terme de réaction diffusion en changeant a et lambda en chaque iteration ')
    global imag_result
    global imag_initial1
    global dt
    global T
    global sigma
    chaleur_win1()
    a=0.5*imag_initial1
    K=int(T/dt)
    psnr=np.zeros(K)
    Lambda = llambda(imag_initial1)
    B=imag_initial1
    for j in range(K):
        B=B+dt*cngradm(B,sigma,Lambda)+dt*f(B,2,a)
        a=a+dt*cngrad(a,Lambda)
        Lambda=llambda(B)
        psnr[j]=PSNR(imag_initial1,B)
    tab=np.array(B*255,dtype=np.uint8) 
    photo=tab_photo(tab)
    imag_result=tab
    Affiche(photo)
    label.config(text='Pennora_M_C réaction 3')
    tracer(psnr)   
   
    
    
    
    














# Utilisation d'un dictionnaire pour conserver une référence
gifdict={}

# 127

Mafenetre.mainloop()
```
